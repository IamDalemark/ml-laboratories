##  Phase 1: Exploratory Data Analysis (EDA)

### 1.1 Data Inspection
- Load the dataset (`pd.read_csv`).
- Check the structure, shape, and data types (`df.info()`, `df.describe()`).
- Indentify the target variable and explicitly state that it is a continuous numeric variable, defining this as a **regression** problem.

### 1.2 Data Cleaning
- **Duplicates**: Check and remove duplicate rows.
- **Missing Values**: 
  - < 5% missing: Drop rows.
  - 5–30% missing: Fill with median (for skewed numerical features) or mode (for categorical).
  - > 30% missing: Consider dropping the column entirely.
- **Target Constraint**: Ensure the target variable doesn't have invalid values (e.g., zero or negative values for prices).

### 1.3 Outlier Detection & Handling
- Outliers heavily impact distance-based and linear regression models.
- **Visuals**: Use boxplots to visualize extreme outliers.
- **IQR Method**: Calculate Q1, Q3, and IQR to identify bounds.
- **Action**: Either clip them (Winsorization) or remove them outright.

### 1.4 Feature Engineering
- **Transformations**: Log-transform highly skewed features (including the target if applicable) to make distributions more normal.
- **Domain Features**: Create meaningful ratio features (e.g., `price_per_sqft`).
- **Date/Time**: Extract Year, Month, Day, and apply cyclical encoding (`sin`/`cos`) for time patterns.
- **Encoding Targets**: Consider Target Encoding for high-cardinality categorical variables.

### 1.5 Visualization
- **Target Distribution**: Always plot a histogram of the target variable first.
- **Skewness**: Check numerical features for skewness (`df.skew()`).
- **Correlations**: Plot a heatmap (e.g., `sns.heatmap`) to identify multicollinearity between predictors and strong correlations with the target.
- **Scatter Plots**: Compare important features directly against the target variable.



## Phase 2: Preprocessing

### 2.1 Categorical Encoding
- **Binary/Ordinal**: Use `LabelEncoder` or manual mapping (e.g., Male=1, Female=0).
- **Nominal (No inherent order)**: Use `OneHotEncoder` / `pd.get_dummies(drop_first=True)` to convert categorical strings to numeric flags.

### 2.2 Train-Test Split
- Split the dataset into `X` (features) and `y` (target).
- **⚠️ CRITICAL**: Do **NOT** use `stratify=y` for regression problems (stratify is strictly for classification).

### 2.3 Feature Scaling
- Essential for distance-based models (KNN, SVR) and linear algorithms (Linear Regression, Ridge, Lasso) to perform correctly.
- Options:
  - **StandardScaler**: Good for normally distributed data.
  - **RobustScaler**: Better choice if your data has substantial remaining outliers.
- **Important Rule**: `fit_transform` on the training set, but only `transform` the test set to avoid data leakage!



##  Phase 3: Model Training

Train at least 3 distinct regression algorithms to compare performance. Typical options include:

1. **Linear Models**:
   - `LinearRegression`: Good baseline.
   - `Ridge` (L2 penalty) / `Lasso` (L1 penalty): Useful when features are correlated.
2. **Distance-Based**:
   - `KNeighborsRegressor` (KNN)
   - `SVR` (Support Vector Regressor) 
3. **Tree-Based / Ensemble**:
   - `DecisionTreeRegressor`
   - `RandomForestRegressor`
   - `GradientBoostingRegressor` / `XGBRegressor` / `LGBMRegressor`



## Phase 4: Hyperparameter Tuning

Improve your best-performing models by exploring hyperparameter spaces:
- Use `GridSearchCV` for exhaustive search or `RandomizedSearchCV` for faster, randomized searches across a parameter grid.
- Focus on tuning key aspects (e.g., `max_depth` and `n_estimators` for Random Forests, or tree-based regularizations).


## Phase 5: Model Evaluation & Comparison

Unlike classification, regression relies on calculating the difference between the actual and predicted quantities. Compare your tuned models using:

- **RMSE (Root Mean Squared Error)**: The standard metric; penalizes larger errors. Lower is better.
- **MAE (Mean Absolute Error)**: Less sensitive to large outliers, representing absolute "off-by" amount.
- **R² Score**: Explains the percentage of variance your model captures (1.0 is perfect, 0 means no better than predicting the mean).
- **RMSLE (Root Mean Squared Log Error)**: Highly effective when evaluating variables spanning large scales (like price distributions) and penalizes relative/percentage errors uniformly.
